```python
import pandas as pd

path_data = 'C:\\data\\ExchangeRate_month.xlsx'
data = pd.read_excel(path_data)
data
```

```python
dataFinal = data.iloc[2:4, 1:]
dataFinal
```

```python
listTmp = list(dataFinal.columns)
listTmp[0] = '구분'
listTmp[1] = '202301'
listTmp[2] = '202302'
listTmp[3] = '202303'
listTmp[4] = '202304'
listTmp[5] = '202305'
listTmp[6] = '202306'
listTmp[7] = '202307'
listTmp[8] = '202308'
listTmp[9] = '202309'
listTmp[10] = '202310'
listTmp[11] = '202311'
listTmp[12] = '202312'
listTmp[13] = '202401'
listTmp[14] = '202402'
listTmp[15] = '202403'
listTmp[16] = '202404'
listTmp[17] = '202405'
listTmp[18] = '202406'
dataFinal.columns = listTmp
dataFinal
```

```python
dataFinal = dataFinal.set_index('구분')
dataFinal
```

```python
dataFinal = dataFinal.T
dataFinal
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

```python
dataFinal['환율'] = dataFinal['환율'].str.replace(',','').astype(float)
dataFinal['절상율(%)'] = dataFinal['절상율(%)'].astype(float)

dataFinal.plot(figsize=(10,8))
plt.title('원/달러 추이')
plt.legend(loc='center left', bbox_to_anchor=(0.7, 0.5))
plt.show()
```

```python
dataLR = dataFinal
dataLR['year'] = dataLR.index
dataLR
```

```python
X = dataLR.iloc[:, -1:]
X
```

```python
y = dataLR.iloc[:, 0]
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split

lr = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2, test_size=0.3)
lr.fit(X_train, y_train)
print('훈령세트점수: {:.2f}'.format(lr.score(X_train, y_train)))
print('테스트 세트 점수: {:3f}'.format(lr.score(X_test, y_test)))
y_pred = lr.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split

for k in range(1, 7):
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, y_train)
    print('train R^2 : {:.3f}'.format(knn.score(X_train, y_train)))
    print('test k : {}, R^2 : {:.3f}'.format(k, knn.score(X_test, y_test)))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

training_accuracy=[]
test_accuracy = []
neighbors=range(1, 7)
for k in neighbors:
    knn=KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, y_train)
    training_accuracy.append(knn.score(X_train, y_train))
    test_accuracy.append(knn.score(X_test, y_test))
    
plt.figure(figsize=(10,8))
plt.plot(neighbors, training_accuracy, label="Training Accuracy")
plt.plot(neighbors, test_accuracy, label='Test Accuracy')
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.xticks(neighbors)
plt.legend()
```

```python
knn = KNeighborsRegressor(n_neighbors=2)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```
